
# Làm sạch dữ liệu khách hàng (Customer Call List)

Notebook này thực hiện các thao tác **data cleaning** trên tập dữ liệu chứa danh sách khách hàng (`Customer Call List.xlsx`).  
Mục tiêu là chuẩn hóa, loại bỏ trùng lặp và làm sạch dữ liệu trước khi sử dụng cho các phân tích tiếp theo.


```python
# Import thư viện pandas để thao tác dữ liệu.
import pandas as pd
```

```python
# Đọc dữ liệu từ file Excel vào DataFrame `df`.
df = pd.read_excel(r"C:\Users\MaiMinhQuan\Desktop\Đồ án tốt nghiệp\Report TV\Report 15-04-2025\Source code\Customer Call List.xlsx")

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```

```python
# Loại bỏ các dòng bị trùng lặp trong dữ liệu.
df = df.drop_duplicates()

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```

```python
# Loại bỏ cột không cần thiết khỏi dữ liệu
df = df.drop(columns = "Not_Useful_Column")  
# Xoá cột có tên là "Not_Useful_Column" khỏi DataFrame
# Đây là cột không chứa thông tin hữu ích cho phân tích, nên được loại bỏ để làm gọn dữ liệu

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```

```python
# Làm sạch dữ liệu trong cột họ (Last_Name)

# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# Xoá dấu "..." ở đầu chuỗi (nếu có) trong cột Last_Name

# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# Xoá ký tự "/" ở đầu chuỗi (nếu có) trong cột Last_Name

# df["Last_Name"] = df["Last_Name"].str.rstrip("_")
# Xoá dấu gạch dưới "_" ở cuối chuỗi (nếu có) trong cột Last_Name

df["Last_Name"] = df["Last_Name"].str.strip("123._/")
# Xoá tất cả các ký tự không mong muốn như: "1", "2", "3", ".", "_", "/"
# `str.strip()` sẽ xoá ký tự ở đầu và cuối chuỗi, không ảnh hưởng đến phần giữa

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```

```python
# Chuẩn hoá định dạng số điện thoại và loại bỏ dữ liệu không hợp lệ

# df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
# Xoá tất cả các ký tự không phải chữ cái hoặc số (ví dụ như dấu "-", "/", "|",...) khỏi chuỗi số điện thoại

# df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# Định dạng lại số điện thoại theo chuẩn "XXX-XXX-XXXX"

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
# Chuyển tất cả giá trị trong cột số điện thoại thành kiểu chuỗi (string), phòng trường hợp bị `NaN`

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# Gán lại định dạng số điện thoại sau khi đã chuyển thành string

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
# Xoá chuỗi không hợp lệ "nan--" khỏi dữ liệu số điện thoại 

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
# Xóa chuỗi `"Na--"` không hợp lệ khỏi dữ liệu số điện thoại

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```

```python
# Tách cột địa chỉ đầy đủ thành các phần nhỏ hơn: địa chỉ đường, bang, mã vùng

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', 2, expand=True)
# Sử dụng hàm `str.split()` để chia chuỗi trong cột `Address` thành 3 phần, ngăn cách bởi dấu phẩy `,`
# Gán các cột kết quả vào ba cột mới: Street_Address (địa chỉ), State (bang/tỉnh), Zip_Code (mã bưu điện)

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```

```python
# Chuẩn hoá giá trị trong cột "Do_Not_Contact" về dạng viết tắt để thống nhất dữ liệu

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
# Thay thế tất cả giá trị "Yes" trong cột bằng "Y"

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
# Thay thế tất cả giá trị "No" trong cột bằng "N"

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```

```python
# Làm sạch các giá trị bị thiếu hoặc không hợp lệ trong dữ liệu

# df = df.replace('N/a','')
# Thay thế các giá trị chuỗi "N/a" (viết tắt của Not available) bằng chuỗi rỗng ''

# df = df.replace('NaN','')
# Thay thế chuỗi "NaN" (chữ, không phải giá trị NaN thực sự) bằng chuỗi rỗng

df = df.fillna('')
# Điền giá trị rỗng '' vào tất cả các ô có giá trị NaN (giá trị thiếu thực sự trong pandas)

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```

```python
# Loại bỏ các dòng dữ liệu của khách hàng đã đánh dấu "Do_Not_Contact" = 'Y'

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        # Kiểm tra nếu giá trị tại cột "Do_Not_Contact" là 'Y' (tức là khách hàng không muốn bị liên hệ)
        df.drop(x, inplace=True)
        # Xoá dòng tương ứng khỏi DataFrame

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```

```python
# Loại bỏ các dòng dữ liệu mà khách hàng không có số điện thoại (Phone_Number bị rỗng)

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        # Kiểm tra nếu ô số điện thoại tại dòng `x` là chuỗi rỗng (tức là không có số)

        df.drop(x, inplace=True)
        # Xoá dòng đó khỏi DataFrame vì không thể liên hệ với khách hàng này

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```

```python
# Đặt lại chỉ số (index) của DataFrame sau khi đã xoá nhiều dòng

df = df.reset_index(drop=True)
# reset_index(): tạo lại chỉ số (index) tuần tự từ 0 cho đến hết

# Hiển thị toàn bộ nội dung bảng dữ liệu df
df
```
